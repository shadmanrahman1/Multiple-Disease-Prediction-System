```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinson_df = pd.read_csv('parkinsons.data')
```

```python
parkinson_df.head()
```

```python
parkinson_df.shape
```

```python
parkinson_df.describe()
```

```python
parkinson_df['status'].value_counts()
```

```python
parkinson_df.groupby('status').mean(numeric_only=True)
```

```python
y_target = parkinson_df['status']

X = parkinson_df.drop(columns='status')
```

TRAIN TEST SPLIT

```python
xtrain, xtest , ytrain , ytest  =  train_test_split(X,y_target,test_size=0.2,random_state=42,stratify=y_target)
```

```python
print(xtrain.shape,xtest.shape,X.shape)
```

PIPELINE


```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
numeric_columns = xtrain.select_dtypes(include='number').columns

scaler_pipeline = ColumnTransformer(
    transformers= [
        ('num',StandardScaler(),numeric_columns)
    ]
)
```

```python
x_train = scaler_pipeline.fit_transform(xtrain)
```

```python
x_test = scaler_pipeline.transform(xtest)
```

```python
x_test.shape
```

Training the Model

```python
model = svm.SVC(kernel='rbf')
```

```python
model.fit(x_train,ytrain)
```

```python
X_train_prediction = model.predict(x_train)
train_data_accuracy = accuracy_score(ytrain,X_train_prediction)
print(train_data_accuracy)
```

```python
y_pred = model.predict(x_test)
test_data_accuracy = accuracy_score(y_pred,ytest)

print(test_data_accuracy)
```

Prediction system

```python
input_data = (88.33300,112.24000,84.07200,0.00505,0.00006,0.00254,0.00330,0.00763,0.02143,0.19700,0.01079,0.01342,0.01892,0.03237,0.01166,21.11800,0.611137,0.776156,-5.249770,0.391002,2.407313,0.249740)

input_df = pd.DataFrame([input_data],columns=X.drop('name',axis=1).columns)

scaled_data1 = scaler_pipeline.transform(input_df) 

prediction = model.predict(scaled_data1)

if prediction[0] == 0:
    print("No Parkinson")
    
else:
    print("Has Parkinson")
```
