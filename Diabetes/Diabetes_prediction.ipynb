Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn import svm

from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND ANALYSIS 

PIMA Diabetes Dataset

```python
diabetes_df = pd.read_csv('diabetes.csv')
```

```python
diabetes_df.groupby('Outcome').mean()
```

```python
y_target = diabetes_df['Outcome']

X = diabetes_df.drop(columns='Outcome',axis=1)
```

```python
X
```

Train Test Split

```python
x_train, x_test , y_train, y_test = train_test_split(X,y_target,test_size=0.2,random_state=2,stratify=y_target)
```

Training Model

```python
from xgboost import XGBClassifier

from sklearn.ensemble import RandomForestClassifier
```

```python
xgb = XGBClassifier(use_label_encoder= False,eval_metric = 'logloss',random_state = 42)
random_forest = RandomForestClassifier(random_state=42)
```

```python
xgb.fit(x_train,y_train)

random_forest.fit(x_train,y_train)
```

Evaluation

```python
from sklearn.metrics import roc_auc_score,confusion_matrix
```

```python
def evaluate_model(model,xtest,ytest):
    y_pred = model.predict(xtest)
    print("Accuracy:",accuracy_score(ytest,y_pred))
    print("ROC Auc:",roc_auc_score(ytest,y_pred))
    print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
    
    
print("XGBClassifier: ")
evaluate_model(xgb,x_test,y_test)

print("\nRandom_Forest:")
evaluate_model(random_forest,x_test,y_test)
```

Making Prediction System

```python
input_data = (10,168,74,0,0,38,0.537,34)
input_df = pd.DataFrame([input_data],columns=X.columns)


#prediction = xgb.predict(input_df)

prediction = random_forest.predict(input_df)

if prediction[0] == 0:
    print("The person is not Diabetic")
    
else:
    print("The person is Diabetic")
```
